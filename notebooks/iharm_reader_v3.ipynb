## How to Use the Modular iharm.jl Module

The iharm.jl module has been designed to be flexible and user-controlled. The iharm.jl gives you full control over loading and managing simulation data directly from your notebook.
The core idea is a simple, explicit, two-step process:

1. Load Data: You tell the module exactly which simulation file to load.

2. Create Accessor: You create an "accessor" dictionary (p) from the loaded data to interact with it.

## Usage

The following code block demonstrates the standard workflow for loading and accessing GRMHD simulation data within a Jupyter notebook.

### Step 1: Define the path to your simulation dump file.
You have the freedom to point this to any .h5 file.

include("/home/usr/Jipole/src/main.jl");
dump_filepath = "/home/usr/Jipole/src/models/iharm3dDumps/dump_001.h5";

### Step 2: Load the simulation data into a dedicated object.
The load_data function reads the HDF5 file and returns an IharmData object.
 
simulation_data = iharm.load_data(dump_filepath)

### Step 3: Create the accessor dictionary 'p' from the data object.
This dictionary contains functions for easy access to each physical variable.

p = iharm.get_p(simulation_data)

### Step 4: Use the accessor 'p' to query and analyze the data.
From this point on, the usage is the same as before.

#### 4.1 Get the value of a single data point (e.g., RHO at grid indices 10, 32, 16)
density_at_point = p[iharm.RHO](10, 32, 16)
println("Density at point (10, 32, 16): ", density_at_point)

#### 4.2 Get a 2D slice of data (e.g., internal energy UU at the j=32 plane)
density_slice = p[iharm.RHO](:, 32, :)
